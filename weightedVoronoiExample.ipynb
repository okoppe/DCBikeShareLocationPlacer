```python
import powerDiagram as pow

import pandas as pd
data = pd.read_csv("bike_share_data/cleaned_data.csv")
points = data[['LATITUDE','LONGITUDE']].to_numpy()
```

```python
cells = pow.createLVDiagram(pow.transform(points)) #creates a voronoi diagram, createLVDiagram also takes weights 
#currently it just assigns a random weight to each location between [0.2 - 1]

#Untransform the cells from the computation
cells = pow.undoTransformC(cells)
```

```python
from IPython.display import display
import folium
        
# Create the Folium map
m = folium.Map(location=[38.90927715, -77.07089615], zoom_start=13)

#add the lines from the voronoi diagram
pow.add_line_collection_to_map(m, cells)

#add the locations of the bikeshare locations to the map
pow.add_points_to_map(m, points)

# Display the map explicitly
display(m)
```
